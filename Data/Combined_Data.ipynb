```python
# Importing dependancies
import pandas as pd
```

```python
# Read files
acs = pd.read_csv("Cleaning Data/Cleaned Data/cleaned_acs.csv")
demo = pd.read_csv("Cleaning Data/Cleaned Data/cleaned_demographic.csv")
costco = pd.read_csv("Cleaning Data/Cleaned Data/cleaned_costco.csv")
```

```python
demo = demo.drop('Unnamed: 0', axis=1)
```

```python
costco
```

```python
costco = costco.drop('Unnamed: 0', axis=1)
```

```python
# ACS Demo merge
acs_demo_merge = acs.merge(demo, on='ZIP', how="outer")

# Merged
merged = acs_demo_merge.merge(costco, on='ZIP', how="outer")
```

```python
acs_demo_merge
```

```python
# Dropping Columns
#merged = merged.drop('Unnamed: 0_x', axis=1)
merged = merged.drop('Unnamed: 0', axis=1)
merged = merged.drop('longitude', axis=1)
merged = merged.drop('latitude', axis=1)
merged = merged.drop('CITY_y', axis=1)
merged = merged.drop('STATE_y', axis=1)
merged = merged.drop('ADDRESS', axis=1)
```

```python
# Adding '0' for Zip Codes
merged['ZIP'] = merged['ZIP'].apply(lambda x : str(x).zfill(5))
```

```python
# Renaming Column
merged.rename(columns={"2021-2026_Population:_Compound_Annual_Growth_Rate":"2021_2026_Population:_Compound_Annual_Growth_Rate"}, inplace=True)
```

```python
# Saving the DataFrame as a csv
merged.to_csv('ml_data.csv')
```
