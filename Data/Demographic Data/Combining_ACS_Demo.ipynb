```python
# Importing dependancies
import pandas as pd
```

```python
# Read files
acs = pd.read_csv("cleaned_acs.csv")
demo = pd.read_csv("cleaned_demographic.csv")
```

```python
# Merge Data
acs_demo_merged = demo.merge(acs,on=["ZIP"])
acs_demo_merged = pd.DataFrame(acs_demo_merged)
```

```python
# Dropping index
acs_demo_merged = acs_demo_merged.drop('Unnamed: 0_x', axis=1)
acs_demo_merged = acs_demo_merged.drop('Unnamed: 0_y', axis=1)
```
